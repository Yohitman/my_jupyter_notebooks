```python
#Поключаем библиотеку sympy:
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from sympy import *
import math as m
def fun(x1,x2,a,b,c):
    return m.sqrt((x2-a-5)**2+(x2-c+3)**2)+m.sqrt((x1-x2)**2+x2**2)+m.sqrt((b+5-x1)**2+(c-3)**2)

init_printing()#для графического отображения формул
a,b,c,x1,x2=symbols('a b c x1 x2')#создаем положительные символьные переменные

a = int(input("вветите a : "))
b = int(input("вветите b : "))
c = int(input("вветите c : "))

L=sqrt((x2-a-5)**2+(x2-c+3)**2)+sqrt((x1-x2)**2+x2**2)+sqrt((b+5-x1)**2+(c-3)**2)#задаем целевую функцию
Ld1=diff(L,x1)#берем производную
Ld2=diff(L,x2)#берем производную
xx=solve([Ld1,Ld2],x1,x2)
simplify(xx)
```

```python
print("x1 равен  :", xx[2][0].evalf(3))
print("x2 равен  :", xx[2][1].evalf(3))
```

```python
%matplotlib inline
l1 = [6, xx[2][1].evalf(3), xx[2][0].evalf(3), c]
l2 =[4, xx[2][1].evalf(3), 0, 4]
l3 =[0, xx[2][0].evalf(3)]
l4 =[0, xx[2][0].evalf(3)]
plt.plot(l1, l2, l3, l4)
plt.axis([0, 8, 0, 8]) 
plt.grid()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x1 = x2 = np.arange(-5.0, 10.0, 0.05)
X, Y = np.meshgrid(x1, x2)
zs = np.array([fun(x1,x2,a,b,c) for x1,x2 in zip(np.ravel(X), np.ravel(Y))])
Z = zs.reshape(X.shape)
ax.plot_surface(X, Y, Z)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
plt.show()
```
