# Метод Гаусса для А

```python
from scipy.optimize import*
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

x0 = -10; y0 = 10
i = 0; e = 0.0001
X=[x0]; Y=[y0]; x_, y_ = 0, y0;

def f1 (x, y):
    return 2*pow(x, 2) + pow(y, 2) + 2*x*y

def fx (x):
    return f1(x, y_)

def fy (y):
    return f1(x_, y)

while True:
    x_ = golden(fx, brack=[-10, 10])
    X.append(x_)
    Y.append(Y[-1])
    y_ = golden(fy, brack=[-10, 10])
    Y.append(y_)
    X.append(X[-1])
    if (abs(X[i] - X[i-1]) < e)  and (abs(Y[i] - Y[i-1]) < e):
        break
    i += 1
    #print(i)

_x_ = np.arange(-10, 10, 0.01)
_y_ = np.arange(-1, 10, 0.01)
x, y = np.meshgrid(_x_, _y_)
z = f1(x, y)

cs = plt.contour(x, y, z, 60)
plt.hold(True)
plt.plot(X,Y)
plt.axis('equal')
plt.show()
print('Ans:','x =',X[-1],';','y =',Y[-1],'.')
print('Count iteration:',(len(X)-1)/2)
print('Count univariation optimization:', (len(X)-1))
(sp.Matrix([X,[0 for x in range(0, len(X))]]).transpose())+(sp.Matrix([[0 for y in range(0, len(Y))],Y]).transpose())
```
