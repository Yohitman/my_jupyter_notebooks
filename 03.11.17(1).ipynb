```python
import sympy as sp
import numpy as np
import pylab
from mpl_toolkits.mplot3d import *
from IPython.display import display, Latex
import matplotlib
import matplotlib.pyplot as plt
sp.init_printing()
def _display(prefix, a):
    display(Latex(' '.join(['$$', prefix, '=', sp.printing.latex(a),' \\$$'])))
a, b, c, x1, x2 =sp.symbols('a b c x1 x2', positive=True)
substdict={a:1, b:2, c:7}
L= (sp.sqrt(x1**2 + a**2)+
    sp.sqrt((x2-x1)**2 + 4**2)+
    sp.sqrt((x2-c)**2 + (4-b)**2)
   )
display('L=',L)
Ld1=L.diff(x1)
Ld2=L.diff(x2)
display('dL/dx1',Ld1)
display('dL/dx2',Ld2)
solutions=sp.solve([Ld1,Ld2],[x1,x2])
for no,solution in enumerate(solutions):
    print('Решение',no,': x1,x2=')
    display(solution)
    print('Проверка: x1,x2=')
    display(solution[0].subs(substdict),solution[1].subs(substdict))
#второе решение(х1=1 и х2=5) подходит
%matplotlib inline
fig = plt.figure()
ax = Axes3D(fig)
i = np.arange(0, 2, 0.01)
j = np.arange(4, 6, 0.01)
x1,x2 = np.meshgrid(i,j)
z = sqrt(1+x1**2)+sqrt((x2-x1)**2+16)+sqrt((7-x2)**2+4)
ax.plot_wireframe(x1, x2, z, rstride = 10, cstride = 10)
plt.show()
```
