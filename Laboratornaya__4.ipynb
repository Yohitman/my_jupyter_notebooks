# Метод покоординатного спуска для первой функции #


```python
from scipy.optimize import*
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

x0 = 4
y0 = 4
i = 0
e = 0.0001
X=[x0]
Y=[y0]
while True:

    if i == 0:
        def f (x, y=y0):
            return pow((1.5-x+x*y), 2) + pow((2.25-x+x*pow(y,2)), 2) + pow((2.625-x+x*pow(y,3)), 2)
        def f (x, y=y_):
            return pow((1.5-x+x*y), 2) + pow((2.25-x+x*pow(y,2)), 2) + pow((2.625-x+x*pow(y,3)), 2)
    x_ = golden(f, brack=[0, 10])
    X.append(x_)
    #print(x_)
    def ff (y, x=x_):
        return pow((1.5-x+x*y), 2) + pow((2.25-x+x*pow(y,2)), 2) + pow((2.625-x+x*pow(y,3)), 2)
    y_ = golden(ff, brack=[0, 10])
    Y.append(y_)
    #print(y_)
    if (abs(X[i] - X[i-1]) < e)  and (abs(Y[i] - Y[i-1]) < e):
        break
    i += 1
    #print(i)

_x_ = np.arange(-4, 4, 0.1)
_y_ = np.arange(-4, 4, 0.1)
x, y = np.meshgrid(_x_, _y_)
z = pow((1.5-x+x*y), 2) + pow((2.25-x+x*pow(y,2)), 2) + pow((2.625-x+x*pow(y,3)), 2)

cs = plt.contour(x, y, z, 20)
plt.hold(True)
plt.plot(X,Y)
plt.show()
print('Ответ:','x =',X[-1],';','y =',Y[-1],'.')
print('количество итераций:',len(X)-1)
print('количество одномерных оптимизаций:',2*(len(X)-1))
```

# Метод Хука-Дживса для второй функции #

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import*
import warnings
warnings.filterwarnings("ignore")

x0 = -10
y0 = 10
i = 0
e = 0.0001
X=[x0]
Y=[y0]
while True:

    if i == 0:
        def f (x, y=y0):
            return x**2*abs(np.sin(2*x))+y**2*abs(np.sin(2*y))-(1/(5*x**2+5*y**2+0.2))+5
    else:
        def f (x, y=y_):
            return x**2*abs(np.sin(2*x))+y**2*abs(np.sin(2*y))-(1/(5*x**2+5*y**2+0.2))+5
    x_ = golden(f, brack=[0, 10])
    X.append(x_)

    def ff (y, x=x_):
        return x**2*abs(np.sin(2*x))+y**2*abs(np.sin(2*y))-(1/(5*x**2+5*y**2+0.2))+5
    y_ = golden(ff, brack=[0, 10])
    Y.append(y_)

    if (abs(X[i] - X[i-1]) < e)  and (abs(Y[i] - Y[i-1]) < e):
        break
    def fff(b, x=x_, y=y_):
        return pow((X[i] + b * (X[i]-  X[i - 1])), 2) + pow(((Y[i] + b * (Y[i] - Y[i - 1])) - 3), 2)+(X[i] + b * (X[i]-  X[i - 1]))*(Y[i] + b * (Y[i] - Y[i - 1]))/2
    b = golden(fff, brack=[0, 10])
    x_ = X[i] + b * (X[i]-  X[i - 1])
    y_ = Y[i] + b * (Y[i] - Y[i - 1])
    i += 1

_x_ = np.arange(-10, 10, 0.1)
_y_ = np.arange(-10, 10, 0.1)
x, y = np.meshgrid(_x_, _y_)
z = x**2*abs(np.sin(2*x))+y**2*abs(np.sin(2*y))-(1/(5*x**2+5*y**2+0.2))+5

cs = plt.contour(x, y, z, 50)
plt.hold(True)
plt.plot(X,Y)
plt.show()
print('Ответ:','x =',X[-1],';','y =',Y[-1],'.')
print('количество итераций:',len(X)-1)
print('количество одномерных оптимизаций:',3*(len(X)-1)-1)
```
